```python
from matplotlib import pyplot as plt
import tensorflow as tf
import os
from PIL import Image
IMAGE_SIZE = (12, 8)
PATH_TO_TEST_IMAGES_DIR = 'test_images/oocl_images'
TEST_IMAGE_PATHS = [ os.path.join(PATH_TO_TEST_IMAGES_DIR, '0708selction{}.jpg'.format(i)) for i in range(19, 25) ]
for image_path in TEST_IMAGE_PATHS:
    image = tf.gfile.FastGFile(image_path, 'r').read()
    with tf.Session() as sess:  
        img_data_jpg = tf.image.decode_jpeg(image)  
        img_data_jpg = tf.image.convert_image_dtype(img_data_jpg, dtype=tf.float32)  
        img_1 = tf.image.flip_left_right(img_data_jpg) 
        plt.figure(figsize=IMAGE_SIZE)
        plt.imshow(img_1.eval())
        plt.show()
```
